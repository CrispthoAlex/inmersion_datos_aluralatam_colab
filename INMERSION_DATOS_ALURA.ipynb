### 2. Precio del m2 por barrio y hacer el gráfico más adecuado para esta nueva variable.

```python
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon'] / inmuebles['Area']
inmuebles.head()
```

```python
# get mean value by 'Barrio'
inmuebles.groupby('Barrio').mean()
```

```python
# get mean value by 'Barrio', adding area properties by 'Barrio'
data_barrios = inmuebles.groupby('Barrio').sum()
data_barrios
```

```python
# get value m2 by 'Barrio'
data_barrios['Valor_m2_Barrio'] = data_barrios['Precio_Millon']/ data_barrios['Area']
data_barrios
```

```python
# create a dictionary with area mean value by 'Barrio'
m2_barrio = dict(data_barrios['Valor_m2_Barrio'])
```

```python
# assign to 'Valor_m2_Barrio' the 'Barrio'
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio']
```

```python
# mapping values from m2_Barrio dictionary to 'Valor_m2_Barrio' column into inmuebles data set
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles['Valor_m2_Barrio']
```

```python
# now we get a 'Valor_m2_Barrio' that have the mean value adding properties by 'Barrio'
inmuebles.head(5)
```

```python
# checking count by 'Barrio'
inmuebles['Barrio'].value_counts()
```

```python
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
top_barrios
```

```python
# reset index
data_barrios.reset_index(inplace= True)
data_barrios
```

```python
# top ten as data frame
data_barrios_query = data_barrios.query('Barrio in @top_barrios')
data_barrios_query
```

```python
# Bar plot
# 'Barrio' with most properties in it

plt.figure(figsize=(10, 8))
# setting axis
ax = sns.barplot(x = 'Barrio', y = 'Valor_m2_Barrio', data = data_barrios_query)
ax.tick_params(axis= 'x', rotation= 45) # legend in axis x
```
