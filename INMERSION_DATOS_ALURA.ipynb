```python
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon'] / inmuebles['Area']
inmuebles.head()
```

```python
# get mean value by 'Barrio', adding area properties by 'Barrio'
data_barrios = inmuebles.groupby('Barrio').sum()
data_barrios
```

```python
# get value m2 by 'Barrio'
data_barrios['Valor_m2_Barrio'] = data_barrios['Precio_Millon']/ data_barrios['Area']
data_barrios
```

```python
# create a dictionary with area mean value by 'Barrio'
m2_barrio = dict(data_barrios['Valor_m2_Barrio'])
```

```python
# assign to 'Valor_m2_Barrio' the 'Barrio'
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio']
```

```python
# mapping values from m2_Barrio dictionary to 'Valor_m2_Barrio' column into inmuebles data set
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles['Valor_m2_Barrio']
```

### Training and Prediction

```python
# Using ML packages
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = data_ml[['COD_UPZ_GRUPO']]
y = data_ml['Precio_Millon']
# split the result. random_state argument, semi-random
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=99)
```
